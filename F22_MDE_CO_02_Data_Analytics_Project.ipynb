## Introduction and Background  of the Study

The Newyork City (NYC) yellow taxi is a organization owning several thousands of taxi cabs in Newyork city USA. These taxis are the only vehicles authorized to pick up and transport street-hailing and prearranged passengers anywhere in the city (see https://www.nyc.gov/site/tlc/businesses/yellow-cab.page). The dataset presented in the analysis and modelling carried out in the Report below is one consisting of features that capture information about single taxi trips in Newyork city.The fare amount and other features were provided to the NYC Taxi and Limousine Commission (TLC) by technology providers authorized under the Taxicab & Livery Passenger Enhancement Programs (TPEP/LPEP).

## Task/Goals

The objective of this study is to examine the historical data in the New York Yellow Taxi dataset to identify relationships between variables and determine if a predictive model can be created to predict fare amounts. The results can provide valuable insights to users of the yellow taxi service and potentially generate new revenue opportunities for the taxi company, such as offering a fare prediction service through a mobile application.

```python
# Check to see how the direction of the trip relates to the amount of the fair charge

# function to select polygons within a provided bounding box range
def select_within_boundingbox(df, BB):
    return (df.pickup_longitude >= BB[0]) & (df.pickup_longitude <= BB[1]) & \
        (df.pickup_latitude >= BB[2]) & (df.pickup_latitude <= BB[3]) & \
        (df.dropoff_longitude >= BB[0]) & (df.dropoff_longitude <= BB[1]) & \
        (df.dropoff_latitude >= BB[2]) & (df.dropoff_latitude <= BB[3])

taxi_trips['delta_longitude'] = taxi_trips.pickup_longitude - taxi_trips.dropoff_longitude
taxi_trips['delta_latitude'] = taxi_trips.pickup_latitude - taxi_trips.dropoff_latitude

# Select trips in Manhattan
BB_manhattan = (-74.025, -73.925, 40.7, 40.8)
pred_manhattan = select_within_boundingbox(taxi_trips, BB_manhattan)

plt.figure(figsize=(15,7))
plt.scatter(taxi_trips[pred_manhattan].delta_longitude, taxi_trips[pred_manhattan].delta_latitude, s=2, alpha=1.0,
            c=np.log1p(taxi_trips[pred_manhattan].fare_amount), cmap='Reds')
plt.colorbar()
plt.xlabel('pickup_longitude - dropoff_longitude')
plt.ylabel('pickup_latitude - dropoff_latidue')
plt.title('log1p(fare_amount)');
```

The section below visualizes how the fare amount  varies with  pick up location. To do that we can calculate the distance from the pick-up location using the pick-up latitude and longitude values to the NYC city center

```python
# Gotten from https://stackoverflow.com/questions/27928/
# Function to calculate the distance between two longitude - latitude points using haversine-formula

def distance(lat_1, lon_1, lat_2, lon_2):
    p = 0.017453292519943295 # Pi/180
    a = 0.5 - np.cos((lat_2 - lat_1) * p)/2 + np.cos(lat_1 * p) * np.cos(lat_2 * p) * (1 - np.cos((lon_2 - lon_1) * p)) / 2
    return 0.6213712 * 12742 * np.arcsin(np.sqrt(a))

nyc = (-74.0063889, 40.7141667) # Longitude and Latitude pair for Nyc center

# add new column to dataframe with distance in mile to the city center
taxi_trips['dist_center'] = distance(nyc[1], nyc[0], taxi_trips.pickup_latitude, taxi_trips.pickup_longitude)
```

Next, we define some functions for visualizing the prediction errors using RMSE

```python
# Function for visualizing the Rmse, evs and a histogram plot of the prediction errors
def plot_prediction_analysis(y, y_pred, figsize=(10,4), title=''):
    fig, axs = plt.subplots(1, 2, figsize=figsize)
    axs[0].scatter(y, y_pred)
    mn = min(np.min(y), np.min(y_pred))
    mx = max(np.max(y), np.max(y_pred))
    axs[0].plot([mn, mx], [mn, mx], c='red')
    axs[0].set_xlabel('$y$')
    axs[0].set_ylabel('$\hat{y}$')
    rmse = np.sqrt(mean_squared_error(y, y_pred))
    evs = explained_variance_score(y, y_pred)
    axs[0].set_title('rmse = {:.2f}, evs = {:.2f}'.format(rmse, evs))

    axs[1].hist(y-y_pred, bins=50)
    avg = np.mean(y-y_pred)
    std = np.std(y-y_pred)
    axs[1].set_xlabel('$y - \hat{y}$')
    axs[1].set_title('Histrogram prediction error, $\mu$ = {:.2f}, $\sigma$ = {:.2f}'.format(avg, std))

    if title!='':
        fig.suptitle(title)
```

```python
plot_prediction_analysis(y_train, y_train_pred, title='Linear Model - Trainingset')
```

To achieve this, another good error metric for evaluating our baseline model is the Mean Absolute Percentage error (MAPE) defined by the formula;



![Mape Formula](images/mape_formula.png)

where

M = Mean Absolute Percentage Error
n = The number of times the summation occurred
At = Actual value
Ft = Forecast value

MAPE is one of the most popular methods for assessing the performance of machine learning forecasting models.

```python
# Define function for computing MAPE
def mape(Y_actual,Y_Predicted):
    Y_actual = np.array(Y_actual)
    Y_Predicted = np.array(Y_Predicted)
    mape = []
    mape.append((np.abs((Y_actual - Y_Predicted)/Y_actual))*100)
    return mape
```

```python
# Compute the mape on the test set
mape_errors = mape(y_test, y_test_pred)
mape_errors
```

```python
outliers = []
[outliers.append(x) for x in mape_errors[0] if x >= 50]
print('The number of predictions made with MAPE above 50% reasonable threshold is {} out of {} predictions made on the test set '.format(len(outliers), len(y_test)))
```

As seen from the result, we have developed a satisfactory model that only produces predictions with errors greater than or equal to 50% 3% of the time.

#### Final feature selection

In this final feature selection approach we utilize sklearn's feature selection library passing our random forest model as the estimator to use in this process. The resulting feature space in this process will then be used for training a final model.

If the model trained on this reduced feature space outperforms the current random forest model we will proceed with it as the best model from the experiment and if it does not perform better it would mean that in our attempt to select the best features we have lost valuable information offered from the features filtered out. In the latter case, we would revert back to the previous model as the best from the experiment

We specify the features to select as 11 out of the 13 passed as input to the algorithm

```python
from sklearn.feature_selection import RFE
rfe_selector = RFE(
    estimator=RandomForestRegressor(),
    n_features_to_select= 11,
    step=20,
    verbose=5,
)

rfe_selector.fit(taxi_copy, taxi_copy.fare_amount)
rfe_support = rfe_selector.get_support()
rfe_support
```
